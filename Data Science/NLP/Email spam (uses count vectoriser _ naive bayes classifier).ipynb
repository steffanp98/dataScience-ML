```
# create df with email csv file 
df = pd.read_csv('emails.csv') #read the CSV file
df.head(5)
```

```
#Print the df shape (rows and cols)
df.shape
```

```
#clean data by removing any duplicates 
df.drop_duplicates(inplace = True)
```

```
#Show the number of missing (NAN, NaN, na) data for each column
df.isnull().sum()
```

```
# download NLP stop words 
# stop words = words which have very little meaning, such as "and", "the", "a", "an"
nltk.download('stopwords')
```

```
#tokeninisation 
# remove stopwords 
# remove punction 
# return a clean list of words for analysis
def cleaning_text(text):

# removing punctiation
  nPunc = [char for char in text if char not in string.punctuation]
  nPunc = ''.join(nPunc)
  # cleaning stop words
  clean_words = [word for word in nPunc.split() if word.lower() not in stopwords.words('english')]
  #return clean list of words 
  return clean_words
```

```
#show the list of tokens creted
df['text'].head().apply(cleaning_text)
```

```
#import the count vectoriser - converts a text doc into a matrix of token counts 
from sklearn.feature_extraction.text import CountVectorizer

## Converts strings to integer counts
vectorizer = CountVectorizer(analyzer=cleaning_text)

## Learn a vocabulary dictionary of all tokens in the raw documents.
bow_transformer = vectorizer.fit(df['text'])    


messages_bow = CountVectorizer(analyzer=cleaning_text).fit_transform(df['text'])
```

```
#Split the data into 70% training (X_train & y_train) and 30% testing (X_test & y_test) data sets
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(messages_bow, df['spam'], test_size = 0.3, random_state = 0)
```

```
#Get the shape of messages_bow
messages_bow.shape
```

```
#create and fit the naive bayes classifier 
from sklearn.naive_bayes import MultinomialNB
classifier = MultinomialNB()
classifier.fit(X_train, y_train)
```

```
#Print the predictions
print(classifier.predict(X_train))

#Print the actual values
print(y_train.values)
```

```
# compare classification model on the traning set 
#show confusion matrix & accuracy
from sklearn.metrics import classification_report,confusion_matrix, accuracy_score

pred = classifier.predict(X_train)
print(classification_report(y_train ,pred ))
print('Confusion Matrix: \n',confusion_matrix(y_train,pred))
print()
print('Accuracy: ', accuracy_score(y_train,pred))
```
