```
#!pip install newspaper3k
!pip install gtts 
from newspaper import Article 
import nltk
from gtts import gTTS
import os
```

```
url = input("please enter your articles URL")
article = Article(url)
```

```
#nlp
article.download() #Download the article
article.parse() #Parse the article
nltk.download('punkt') #Download the 'punkt' package
article.nlp() #Apply Natural Language Processing (NLP)
```

```
#get articles text 
mytext = article.text
#print arcticles text
print(mytext)
#language to convert to 
language = 'en' #English
```

```
myReading = gTTS(text=mytext, lang=language, slow=False)

#save converted file into mp3 format 
myReading.save("read_article.mp3")
```

```
#play converted file 
os.system("start read_article.mp3")
```
