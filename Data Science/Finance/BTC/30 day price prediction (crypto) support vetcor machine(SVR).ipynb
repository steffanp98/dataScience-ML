```
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.svm import SVR
```

```
# create and read data from cvs file to df 
df = pd.read_csv('BitcoinPrice.csv')
df.head(5) # diplays the first 5 rows of the df
```

```
#only show the price col 
df.drop(['Date'], 1, inplace=True)
#display new df with the price col
df.head(5)
```

```
#A variable for predicting 'x' days out into the future
prediction_days = 30 # x = 30

#Create another column (the target or dependent variable) shifted 'x' units up
df['Prediction'] = df[['Price']].shift(-prediction_days)
```

```
df.tail(5)
```

```
# x data set (inderpendant set)
#convert df into numyy array for computation 
x = np.array(df.drop(['Prediction'],1))

#remove the x amount of rows where x is the prediction days 
#removing the rows where x = x predicted days 
x = x[:len(df)-prediction_days]
print(x)
```

```
#Y data set inderpendant (Y)
y = np.array(df['Prediction'])
# Get all of the y values except the last 'n' rows
y = y[:-prediction_days]
print(y)
```

```
#split datasets into x_train and y_train (70% train and 30% test)
from sklearn.model_selection import train_test_split
x_train, X_test, y_train, Y_test = train_test_split(x,y, test_size= 0.3)
```

```
#set the preditcions days array = the 30 rows of the price col 
prediction_days_array = np.array(df.drop(['Prediction'],1))[-prediction_days:]
print(prediction_days_array)
```

```
#create & fit the svr (support vector machine) using the radial basis fucntion
# radial basis function = an approximation of a multivariable function. Can be used in more than one dimension
from sklearn.svm import SVR
svr_rbf = SVR(kernel='rbf', C=1e3, gamma=0.00001)
svr_rbf.fit(x_train, y_train)
```

```
# test the acc of the model 
# 1.0 is the most accurate
svr_rbf_confidence = svr_rbf.score(X_test,Y_test)
print("svr accuracy:", svr_rbf_confidence) 
```

```
#show models predction value 
#I.e the prediction of the price of BTC over the next 30 days
svm_predication = svr_rbf.predict(X_test)
print(svm_predication)

print ()

#actual value 
print(Y_test)
```
